## The Guassian Distribution 

### Adapted from https://www.gaussianwaves.com/2020/06/using-matplotlib-histogram-in-python/  

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
#import seaborn as sns
# if using a Jupyter notebook, inlcude:
%matplotlib inline
```

```python
mu=10;sigma=2.5 #mean=10,deviation=2.5
L=100000 #length of the random vector
bins = 100

#Random samples generated using numpy.random.normal()
samples_normal = np.random.normal(loc=mu,scale=sigma,size=(L,1))
```

```python
plt.style.use('ggplot')

fig_norm, ax0 = plt.subplots(figsize=(9,6))
(values, bins, _) = ax0.hist(samples_normal,bins = bins, density=True,label="Histogram of samples") 
#Compute and plot histogram, return the computed values and bins
bin_centers = 0.5*(bins[1:] + bins[:-1])
pdf = norm.pdf(x = bin_centers, loc=mu, scale=sigma) #Compute probability density function
ax0.plot(bin_centers, pdf, label="PDF",color='black') #Plot PDF
ax0.legend()#Legend entries
ax0.set_title('PDF of samples from numpy.random.normal()')
```
